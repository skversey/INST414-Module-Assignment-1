```python
import pandas as pd
```

```python
df = pd. read_csv('Player Per Game.csv')
df.head()
```

```python
filtered_df = df[df['season']== 2024]
filtered_df = filtered_df[df['pts_per_game'] >= 20]
filtered_df = filtered_df[['player', 'pos', 'tm', 'pts_per_game', 'trb_per_game', 'ast_per_game', 'stl_per_game', 'blk_per_game']]
filtered_df.head()
```

```python
ppg_df = filtered_df.sort_values(by=['pts_per_game'], ascending=False)
ppg_df.head()
```

```python
rpg_df = filtered_df.sort_values(by=['trb_per_game'], ascending=False)
rpg_df.head()
```

```python
ast_df = filtered_df.sort_values(by=['ast_per_game'], ascending=False)
ast_df.head()
```
